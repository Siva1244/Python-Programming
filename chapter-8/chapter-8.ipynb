```python
#programming assignment_question-11 and 12
def main():
    inp=input("Enter file name(sai.txt) : ")
    infile=open(inp,"r")
    coolingdays=0
    heatingdays=0
    for i in infile.readlines():
        t=i.split(",")
        for j in t:
            j=int(j)
            if (j>80):
                coolingdays=coolingdays+(j-80)
            elif j<60:
                heatingdays=heatingdays+(60-j)
    print("Heating degree days is {0} and cooling degree days is {1}".format(heatingdays,coolingdays))
    
```

```python
#programming assignment_Question-13
from graphics import *
import math
def main():
    win=GraphWin("Regression",500,500)
    win.setCoords(0,0,100,100)
    win.setBackground("white")
    r=Rectangle(Point(1,1),Point(10,5))
    c=r.getCenter()
    Text(c,"End").draw(win)
    r.draw(win)
    n=0
    sumx=0
    sumy=0
    sxy=0
    sxx=0
    p=win.getMouse()
    minx=p.getX()
    maxx=p.getY()
    while not(p.getX()<=10 and p.getY()<=5):
        p.draw(win)
        n=n+1
        x=p.getX()
        y=p.getY()
        if x<minx:
            minx=x
        if x>maxx:
            maxx=x
        sumx=sumx+x
        sumy=sumy+y
        sxy=sxy+x*y
        sxx=sxx+(x**2)
        p=win.getMouse()
    xbar=sumx/n
    ybar=sumy/n
    m=(sxy-n*xbar*ybar)/(sxx-(n*(xbar**2)))
    x1=minx
    y1=ybar + m*(x1-xbar)
    x2=maxx
    y2=ybar +m*(x2-xbar)
    Line(Point(x1,y1),Point(x2,y2)).draw(win)
    
    win.getMouse()
    win.close()
```
