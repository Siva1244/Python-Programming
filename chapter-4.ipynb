```python
# this will take three points(by clicking) from us and print triangle
from graphics import *
def main():
    win=GraphWin("Draw a Triangle")
    win.setCoords(0.0,0.0,10.0,10.0)
    message=Text(Point(5,0.5),"click on three points")
    message.draw(win)
    
    p1=win.getMouse()
    p1.draw(win)
    p2=win.getMouse()
    p2.draw(win)
    p3=win.getMouse()
    p3.draw(win)
    
    triangle=Polygon(p1,p2,p3)
    triangle.setFill("peachpuff")
    triangle.setOutline("cyan")
    triangle.draw(win)
    
    message.setText("click anywhere to quit.")
    win.getMouse()
```

```python
from graphics import *
def main():
    win = GraphWin("Celsius Converter",400,300)
    win.setCoords(0.0,0.0,3.0,4.0)
    
    Text(Point(1,3),"Celcius Temperature:").draw(win)
    Text(Point(1,1),"Fahrenheit Temperature:").draw(win)
    inputText=Entry(Point(2.25,3),5)
    inputText.setText("0.0")
    inputText.draw(win)
    outputText=Text(Point(2.25,1),"")
    outputText.draw(win)
    button=Text(Point(1.5,2.0),"Convert It")
    button.draw(win)
    Rectangle(Point(1,1.5),Point(2,2.5)).draw(win)
    
    win.getMouse()
    
    celcius = float(inputText.getText())
    fahrenheit = 9.0/5.0 *celcius+32
    
    outputText.setText(round(fahrenheit,2))
    button.setText("Quit")
    
    win.getMouse()
    win.close()
```

```python
# programming assignment-Question-7
from graphics import *
import math
def main():
    win=GraphWin("Circle intercept",400,400)
    win.setCoords(-20.0,-20.0,20.0,20.0)
    
    Text(Point(-10,18),"Radius of circle (0-10): ").draw(win)
    r=Entry(Point(10,18),5)
    r.setText("0.0")
    r.draw(win)
    win.getMouse()
    Text(Point(-10,15),"y intercept (0-10) : ").draw(win)
    y=Entry(Point(10,15),5)
    y.setText("0.0")
    y.draw(win)
    win.getMouse()
    
    yi=float(y.getText())
    ri=float(r.getText())
    
    ci=Circle(Point(0,0),ri)
    ci.setFill("yellow")
    ci.draw(win)
    
    Line(Point(-10,yi),Point(10,yi)).draw(win)
    Line(Point(-10,0),Point(10,0)).draw(win)
    Line(Point(0,-10),Point(0,10)).draw(win)
    Text(Point(-12,0),"(-10,0)").draw(win)
    Text(Point(12,0),"(10,0)").draw(win)
    Text(Point(0,-12),"(0,-10)").draw(win)
    Text(Point(0,12),"(0,10)").draw(win)
    
    win.getMouse()
    
    x=math.sqrt(ri**2-yi**2)
    e=Circle(Point(x,yi),0.5)
    e.setFill("red")
    e.draw(win)
    f=Circle(Point(-x,yi),0.5)
    f.setFill("red")
    f.draw(win)
    win.getMouse()
    
    
    Text(Point(-10,-16),"x intercept is :  ").draw(win)
    Text(Point(8,-16),x).draw(win)
    Text(Point(-10,-18),"x intercept is :  ").draw(win)
    Text(Point(8,-18),-x).draw(win)
    
    win.getMouse()
    win.close()
    
```

```python
# programming assignment-Question-8
from graphics import *
import math
def main():
    win=GraphWin("Line segment",300,300)
    win.setCoords(0.0,0.0,10.0,10.0)
    p1=win.getMouse()
    p2=win.getMouse()
    a=Line(p1,p2)
    a.draw(win)
    win.getMouse()
    mid=a.getCenter()
    x=mid.getX()
    y=mid.getY()
    Text(Point(2,4),"Mid point of line is : ").draw(win)
    Text(Point(7,4),x).draw(win)
    Text(Point(7,3),y).draw(win)
    win.getMouse()
    dx=p1.getX()-p2.getX()
    dy=p1.getY()-p2.getY()
    slope=dy/dx
    Text(Point(2,2),"slope of line is : ").draw(win)
    Text(Point(7,2),slope).draw(win)
    win.getMouse()
    leng=math.sqrt(dx**2+dy**2)
    Text(Point(2,1),"length of line is : ").draw(win)
    Text(Point(7,1),leng).draw(win)
    win.getMouse()
    win.close()
    
```

```python
# programming assignment-Question-10
from graphics import *
import math
def main():
    win=GraphWin("Triangle",400,400)
    win.setCoords(0.0,0.0,10.0,10.0)
    p1=win.getMouse()
    p2=win.getMouse()
    p3=win.getMouse()
    a=Polygon(p1,p2,p3)
    a.draw(win)
    
    win.getMouse()
    
    x1=p1.getX()
    x2=p2.getX()
    x3=p3.getX()
    y1=p1.getY()
    y2=p2.getY()
    y3=p3.getY()
    
    a=math.sqrt((x2-x1)**2+(y2-y1)**2)
    b=math.sqrt((x3-x2)**2+(y3-y2)**2)
    c=math.sqrt((x3-x1)**2+(y3-y1)**2)
    s=(a+b+c)/2
    area=math.sqrt(s*(s-a)*(s-b)*(s-c))
    perm=a+b+c
    
    Text(Point(2,2),"Area of triangle is : ").draw(win)
    Text(Point(7,2),area).draw(win)
    
    win.getMouse()
    
    
    
    Text(Point(3,1),"perimeter of triangle is : ").draw(win)
    Text(Point(7,1),perm).draw(win)
    win.getMouse()
    win.close()
```

```python
# programming assignment-Question-11
from graphics import *
import math
def main():
    win=GraphWin("House",400,400)
    win.setCoords(0.0,0.0,10.0,10.0)
    
    p1=win.getMouse()
    p2=win.getMouse()
    a=Rectangle(p1,p2)
    a.draw(win)
    
    
    t2=Point(p1.getX(),p2.getY())
    
    p3=win.getMouse()
    q=(p1.getX()+(1/6)*(p2.getX()-p1.getX()))
    do=Point(q,p1.getY())
    Rectangle(do,Point(q+(  2*(p3.getX()-q)   ),p3.getY())).draw(win)
    
    
    p4=win.getMouse()
    x=p4.getX()
    y=p4.getY()
    dy=y-p3.getY()
    Rectangle(Point(x-dy,y-dy),Point(x+dy,y+dy)).draw(win)
    
    
    p5=win.getMouse()
    Polygon(p2,t2,p5).draw(win)
    
    win.getMouse()
    win.close()
```
